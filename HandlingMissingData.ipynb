```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.isnull(np.nan)
```

```python
pd.isnull(pd.Series([1, 2, np.nan, 4]))
# 3 + np.nan
# 3 + np.inf
```

```python
pd.notnull(None)
```

```python
arr = pd.DataFrame({
    'A': [1, 2, np.nan], 
    'B': [4, np.nan, np.nan],
    'C': [7, 8, 9]
    })
arr.notnull()
```

```python
pd.notnull(arr).any(axis=1)
```

```python
arr.dropna(how='any')
```

```python
arr.dropna(thresh=2)
```

```python
arr.fillna(method='ffill')
# arr.fillna(method='bfill')
```

```python
x = np.arange(-10, 11)
plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.plot(x , x**2 , label='A')
plt.plot([0,0, 0], [-10, 0, 100])  # Adding red dots at  (0,-10), (0,0), and (0,100)
plt.title('Plotting x^2')
plt.xlabel('x')
plt.ylabel('x^2')

plt.subplot(1,2,2)
plt.plot(x , -1*(x**2) , label='B')
plt.plot([-10,0, 10], [-50, -50, -50])  # Adding red dots at (0,-10), (0,0), and (0,100)
plt.title('Plotting x^3')
plt.xlabel('x')
plt.ylabel('x^3')
```

```python
fig , axes = plt.subplots(figsize=(12,6))
```

```python
fig , axes = plt.subplots(figsize=(12,6))
axes.plot(x, x+0 , '-og', label="solid green")
axes.plot(x, x+1 , '--c', label="dashed cyan")
axes.plot(x, x+2 , '-.b', label="dashed blue")
axes.plot(x, x+3 , ':^r', label="doted red")
```

```python
values = np.random.randn(1000)
plt.subplots(figsize=(12,6))
plt.hist(values , bins = 100 , alpha=0.8 ,
         histtype='bar' , color='steelblue', edgecolor='green')
plt.xlim(xmin= -4, xmax= +4)
plt.show
```

```python
from scipy import stats
density = stats.kde.gaussian_kde(values)
density
```

```python
values2 = np.linspace(min(values)-10, max(values)+10, 100)
plt.subplots(figsize=(12,6))
plt.plot(values2, density(values2), color='orange')
plt.fill_between(values2, 0 ,density(values2), alpha=0.5 , color='orange')
plt.xlim(xmin=-4,xmax=4)
```

```python
values2 = np.linspace(min(values)-10, max(values)+10, 100)
plt.subplots(figsize=(12,6))
plt.hist(values , bins = 100 , alpha=0.8 ,density=1 ,
         histtype='bar' , color='steelblue', edgecolor='green')
plt.plot(values2, density(values2), color='orange')
# plt.fill_between(values2, 0 ,density(values2), alpha=0.5 , color='orange')
plt.xlim(xmin= -4, xmax= +4)
```
