```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
# df['target'] = iris.target
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df['flower_names'] = [iris.target_names[x] for x in df['target']]
# df[44:65]
```

```python
dfs = df[df.target == 0]
dfvers = df[df.target == 1]
dfvirg = df[df.target == 2]
```

```python
plt.scatter(dfs['sepal length (cm)'], dfs['sepal width (cm)'], color='blue', label='setosa')
plt.scatter(dfvers['sepal length (cm)'], dfvers['sepal width (cm)'], color='orange', label='versicolor')
plt.scatter(dfvirg['sepal length (cm)'], dfvirg['sepal width (cm)'], color='green', label='virginica')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Flower Sepal Dimensions')
plt.legend()
```

```python
plt.scatter(dfs['sepal length (cm)'], dfs['petal width (cm)'], color='blue', label='setosa')
plt.scatter(dfvers['sepal length (cm)'], dfvers['petal width (cm)'], color='orange', label='versicolor')
plt.scatter(dfvirg['sepal length (cm)'], dfvirg['petal width (cm)'], color='green', label='virginica')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Flower Sepal Dimensions')
plt.legend()
```

```python
x = df.drop(['target', 'flower_names'], axis=1)
y = df['target']
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
