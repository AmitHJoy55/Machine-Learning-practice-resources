```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin' ,'SibSp' ,'Parch', 'Embarked'],inplace=True)
df.head()
```

```python
target = df['Survived']
input = df.drop(columns=['Survived'], axis=1)
```

```python
dummies = pd.get_dummies(input.Sex )
dummies.head()
input = pd.concat([input, dummies], axis=1)
input.head()
```

```python
input.drop('Sex',inplace=True, axis=1)
input.head()
```

```python
input.columns[input.isna().any()]
```

```python
input.Age = input.Age.fillna(input.Age.mean())
input.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(input,target, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model= GaussianNB()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
model.predict_proba(X_test)
```
